## Machine Learning - K Means Clustering 

K Means Clustering is a process, the overall goal is to divide data into distinct groups such that observations within each group are 

**The k-means algorithm works as follows:**

- Choose the number of clusters K.
- Randomly assign each data point to a cluster.
- Repeat the following steps until the clusters stop changing:

    a. Compute the cluster centroids by taking the mean of the data points in each cluster.
    
    b. Assign each data point to the cluster with the closest centroid.

**Workflow**

- In the visualization, the process starts with observations (data points) 
- In Step 1, each observation is randomly assigned to a cluster, and the cluster centroids are computed. Initially, these centroids overlap because of the random initial assignments
- In Step 2, each observation is then assigned to the nearest centroid. Step 2 is repeatedly performed, leading to new cluster centroids, until no new clusters are formed, meaning data points are not being reassigned to different centroids.

**N: Note that k means clustering is an unsupervised learning algorithm and it tries to cluster data based on their similarities**

### Import Libraries

```python
# starting with importing libariraies 

import numpy as np
import pandas as pd
```

```python
# importing data visualization libraries 

import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

### Get the data 

I will create artificial data with sklearn capabilities.

```python
from sklearn.datasets import make_blobs
```

```python
make_blobs
```

```python
data = make_blobs(n_samples=200, n_features=2,cluster_std=1.8,
                 random_state=101)
```

```python
data
```

### Visualize Data

```python
# trying to plot our data 

plt.scatter(data[0][:,0],data[0][:,1] )
```

```python
data[1]
```

```python
# add it to our original scatter

plt.scatter(data[0][:,0],data[0][:,1],c=data[1], cmap='rainbow' )
```

### Creating the Clusters

using sklearn to creat K Means Cluster

```python
from sklearn.cluster import KMeans
```

```python
# creating a K Means object, remember to know the number of clusters beforehand.

kmeans = KMeans(n_clusters=3) # as it shows in the scatter plot

```

```python
kmeans.fit(data[0])
```

```python
# let's try centers of the clusters method

kmeans.cluster_centers_
```

```python
# I try lables method, which return labels it believes to be true for the clusters.

kmeans.labels_
```

**K Means Clustering is unsupervised, so you wouldn't use it to predict labels you would use it to find labels in your data**

now let's check how K Means compares to our original data by using plotting in order to be able to see it

```python
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True,figsize=(12,8))

ax1.set_title('K Means')
ax1.scatter(data[0][:,0],data[0][:,1])

ax2.set_title("Original")
ax2.scatter(data[0][:,0],data[0][:,1])
```

now let's try to add some arguments to the code, to make it look more comparable 

```python
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True,figsize=(12,8))

# here we take the data from kmeans lables array
ax1.set_title('K Means')
ax1.scatter(data[0][:,0],data[0][:,1],
            c=kmeans.labels_,cmap='rainbow')

# while here we use the orginal array data
ax2.set_title("Original")
ax2.scatter(data[0][:,0],data[0][:,1],
            c=data[1],cmap='rainbow')
```

Once you change the number of clusters in 'kmeans = KMeans(n_clusters=3)', by increasing or decreasing the cluster number, it will change the number of clusters shown in the scatter plot. As per the example, I will redo it with different cluster numbers. 

```python
# Example

kmeans = KMeans(n_clusters=5)

kmeans.fit(data[0])

f, (ax1, ax2) = plt.subplots(1, 2, sharey=True,figsize=(12,8))

# here we take the data from kmeans lables array
ax1.set_title('K Means')
ax1.scatter(data[0][:,0],data[0][:,1],
            c=kmeans.labels_,cmap='rainbow')

# while here we use the orginal array data
ax2.set_title("Original")
ax2.scatter(data[0][:,0],data[0][:,1],
            c=data[1],cmap='rainbow')
```

here we can notice how the plot changed into 5 different clusters to match our input
