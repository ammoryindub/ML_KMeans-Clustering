### K Means Cluster Creation


Now it is time to create the Cluster labels

**Import KMeans from SciKit Learn**

```python
from sklearn.cluster import KMeans
```

```python
# creat an instance of KMenas model with 2 clusters 

kmeans = KMeans(n_clusters=2)
```

```python
# fit the model

kmeans.fit(df.drop('Private',axis=1))
```

#### N: Note that we dont do a train to split here because that wouldn't make sense on an unsupervised learning algorithms as we don't actually have labels to split

```python
# cluster center vectors

kmeans.cluster_centers_
```

### Evaluation

there is no perfect way to evaluate clustering if you don't have the labels

**for this dataset I will create a new column for df called 'Cluster', which is a 1 for a Private school, and a 0 for a public school**

```python
def converter(cluster):
    if cluster=='Yes':
        return 1
    else:
        return 0
```

```python
df['Cluster'] = df['Private'].apply(converter)
```

```python
df.head()

# we should find the 'Cluster' column initiated, shich aligns with 'Private' column.
```

**create a confusion matrix and classification report to see how well the Kmeans clustering worked without being given any labels**

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python

print(confusion_matrix(df['Cluster'],kmeans.labels_))
```

```python
print(classification_report(df['Cluster'],kmeans.labels_))
```
